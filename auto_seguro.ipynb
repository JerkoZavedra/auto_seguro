```python
# Principal

import funciones_auto_seguro as funciones

def mostrar_menu():
  print("Bienvenido a Auto Seguro")
  print("1) Grabar vehiculo")
  print("2) Buscar vehiculo")
  print("3) Imprimir certificados")
  print("4) Salir")

def main():
  while True:
    mostrar_menu()
    opcion = input("Elija una opcion (1/2/3/4): ")

    if opcion == '1':
      funciones.grabar_vehiculo()
    elif opcion == '2':
      funciones.buscar_vehiculo()
    elif opcion == '3':
      funciones.imprimir_certificados()
    elif opcion == '4':
      print("Gracias por usar Auto Seguro.")
      break
    else:
      print("Opcion no valida. Por favor elija una opcion valida (1/2/3/4)")

if __name__ == "__main__":
  main()
```
